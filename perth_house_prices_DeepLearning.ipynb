```python
#importing dependencies

import pandas as pd
import os
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Loading CSV

df = pd.read_csv(os.path.join("Resources", "perth_house_prices.csv"))

df.head()
```

```python
#Selecting Columns of interest

df = df[["BUILD_YEAR","BEDROOMS","BATHROOMS","LAND_AREA","FLOOR_AREA","CBD_DIST","NEAREST_SCH_RANK","PRICE"]]
df
```

```python
#Deleting all empty rows
df = df.dropna()
df
```

```python
# Outputting Dataframe into CSV

df.to_csv('file_name.csv', index=False)
```

```python
# Prepocessing data

X = df[["BUILD_YEAR","BEDROOMS","BATHROOMS","LAND_AREA","FLOOR_AREA","CBD_DIST","NEAREST_SCH_RANK"]]
y = df["PRICE"].values.reshape(-1,1)
print(X.shape, y.shape)
```

```python
# Splitting data into training data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
# Scaling data

from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
```

```python
# Transforming Data

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

Deep Learning

```python
from tensorflow.keras.utils import to_categorical
```

```python
# One-hot encoding
y_train_categorical = to_categorical(y_train_scaled)
y_test_categorical = to_categorical(y_test_scaled)
```

```python
# Creating a normal neural network with 3 inputs, 6 hidden nodes, and 1 output
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=14, activation='relu', input_dim=7))
model.add(Dense(units=14, activation='relu'))
model.add(Dense(units=14, activation='relu'))

model.add(Dense(units=1, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model (adam optimizer)
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=300,
    shuffle=True,
    verbose=2
)
```

```python
# Compile the model (RMSProp optimizer)
model.compile(optimizer='RMSProp',
              loss='mean_squared_error',
              metrics=['accuracy'])
```
